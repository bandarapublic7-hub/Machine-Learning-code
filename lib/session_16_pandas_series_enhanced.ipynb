## 📌 What is Pandas?

**Pandas** is a fast, powerful, flexible, and easy-to-use open-source **data analysis and manipulation** library built on top of Python.

- Built on top of **NumPy** (so it's fast)
- Provides two core data structures: **Series** (1D) and **DataFrame** (2D)
- Essential for any data science / ML workflow

🔗 Official docs: https://pandas.pydata.org/about/index.html


## 📌 What is a Pandas Series?

A **Pandas Series** is a **one-dimensional labeled array** capable of holding any data type (integers, strings, floats, Python objects, etc.).

Think of it as:
- A **single column** in a table / spreadsheet
- Like a Python **list**, but with an **index** attached to every element
- Like a Python **dict**, but vectorized and with many built-in methods

```
Index   Values
  0  →   10
  1  →   20
  2  →   30
```

> 💡 Every Series has an **index** (labels) and **values** (data).


## 1️⃣ Importing Pandas

Always import Pandas as `pd` (convention). NumPy (`np`) is imported because Pandas is built on top of it and many operations use `np.nan`, `np.array`, etc.


````python
import numpy as np
import pandas as pd

print("Pandas version:", pd.__version__)
print("NumPy version:", np.__version__)
````

## 2️⃣ Creating a Series

### 📂 From a Python List

When you pass a plain list, Pandas automatically assigns a **default integer index** starting from 0.


````python
# From a list of strings
country = ['India', 'Pakistan', 'USA', 'Nepal', 'Srilanka']
pd.Series(country)
# Index: 0,1,2,3,4  (auto-assigned)
````

````python
# From a list of integers
runs = [13, 24, 56, 78, 100]
runs_ser = pd.Series(runs)
runs_ser
````

### 🏷️ Custom Index
You can pass your own labels as the `index` parameter. The length of index must match the length of data.


````python
marks = [67, 57, 89, 100]
subjects = ['maths', 'english', 'science', 'hindi']

# Custom index: now each mark is labeled by subject
pd.Series(marks, index=subjects)
````

### 📛 Setting a Name
A Series can have a `.name` attribute — useful when combining multiple Series into a DataFrame later.


````python
marks = pd.Series(marks, index=subjects, name='Nitish ke marks')
marks
````

### 📂 From a Dictionary
When you create a Series from a dict, the **keys become the index** and the **values become the data** automatically.


````python
marks_dict = {
    'maths': 67,
    'english': 57,
    'science': 89,
    'hindi': 100
}

marks_series = pd.Series(marks_dict, name='nitish ke marks')
marks_series
````

### 🔢 From a Scalar Value
You can create a Series from a single value — it gets broadcast to match the index length.


````python
# Same value repeated for each index label
pd.Series(5, index=['a', 'b', 'c', 'd'])
````

### 🔢 From a NumPy Array


````python
arr = np.array([10, 20, 30, 40, 50])
pd.Series(arr, index=['a', 'b', 'c', 'd', 'e'])
````

## 3️⃣ Series Attributes

Attributes give you **metadata** about the Series without performing any computation.

| Attribute | Description |
|-----------|-------------|
| `.size` | Total number of elements (including NaN) |
| `.dtype` | Data type of the values |
| `.name` | Name of the Series |
| `.is_unique` | True if all values are unique |
| `.index` | The index (labels) of the Series |
| `.values` | The underlying NumPy array of values |
| `.shape` | Tuple with dimensions, e.g. `(5,)` |
| `.ndim` | Number of dimensions (always 1 for Series) |
| `.empty` | True if the Series has no elements |


````python
# size — number of elements
print("size:", marks_series.size)

# dtype — data type of values
print("dtype:", marks_series.dtype)

# name
print("name:", marks_series.name)

# is_unique — True if no repeated values
print("is_unique:", marks_series.is_unique)
print("is_unique (with dups):", pd.Series([1, 1, 2, 3]).is_unique)

# index — the labels
print("index:", marks_series.index)

# values — returns underlying numpy array
print("values:", marks_series.values)

# shape and ndim
print("shape:", marks_series.shape)
print("ndim:", marks_series.ndim)

# empty
print("empty:", pd.Series([]).empty)
````

## 5️⃣ Series Methods — Exploration

````python
s = pd.Series(range(1, 21))  # 1 to 20

print("head(5):")
print(s.head())       # first 5

print("
head(3):")
print(s.head(3))      # first 3

print("
tail(5):")
print(s.tail())       # last 5

print("
tail(3):")
print(s.tail(3))      # last 3
````

Counts how many times each **unique value** appears. Results are sorted by frequency (most common first).
Very useful for categorical data.

### 🔢 sort_values() and sort_index()
- `sort_values()` — sorts by the **data values**
- `sort_index()` — sorts by the **index labels**
- `ascending=False` for descending order
- `inplace=True` modifies the original Series instead of returning a new one


````python
nums = pd.Series([40, 10, 30, 20, 50], index=['e','a','c','b','d'])

print("sort_values (asc):")
print(nums.sort_values())

print("
sort_values (desc):")
print(nums.sort_values(ascending=False))

print("
sort_index:")
print(nums.sort_index())
````

## 6️⃣ Series — Mathematical Methods

These methods work on **numeric Series** and help you quickly summarize data.

| Method | Description |
|--------|-------------|

## 7️⃣ Series Indexing

Pandas has two main indexers:
- **`[]`** — label-based or integer-based depending on context (can be confusing!)
- **`.iloc[]`** — always **integer position-based** (like Python list indexing)
- **`.loc[]`** — always **label-based**

> ✅ Best practice: prefer `.iloc` and `.loc` over `[]` to avoid ambiguity.


````python
x = pd.Series([12, 13, 14, 35, 46, 57, 58, 79, 9])

# Integer indexing
print("x[0]:", x[0])
print("x[3]:", x[3])
````

````python
# iloc — position-based (always safe for integer index)
print("x.iloc[0]:", x.iloc[0])
print("x.iloc[-1]:", x.iloc[-1])   # last element
````

````python
# loc — label-based
marks_series.loc['maths']        # access by label
# marks_series.loc['maths':'science']  # slice by label (inclusive on both ends!)
````

````python
# Slicing — vk[5:16]
x[2:6]   # elements at positions 2,3,4,5
````

````python
# Negative slicing — last 5 elements
x[-5:]
````

````python
# Step slicing — every 2nd element
x[::2]
````

````python
# Fancy indexing — pass a list of positions
x[[0, 2, 4, 6]]
````

````python
# Label-based fancy indexing
marks_series[['maths', 'science']]
````

## 8️⃣ Editing a Series

You can **modify values** using the same indexing syntax. If you assign to a **label that doesn't exist**, Pandas will **add a new element**.


````python
s = pd.Series([67, 57, 89, 100], index=['maths','english','science','hindi'], name='marks')

# Modify by position
s[1] = 60
print("After s[1] = 60:")
print(s)
````

````python
# Add a new subject (new index label)
s['evs'] = 95
print("After adding 'evs':")
print(s)
````

````python
# Modify using slicing
s2 = pd.Series([10, 20, 30, 40, 50])
s2[1:3] = [200, 300]
print("After s2[1:3] = [200,300]:")
print(s2)
````

````python
# Modify using fancy indexing
s2[[0, 4]] = [0, 0]
print("After s2[[0,4]] = [0,0]:")
print(s2)
````

## 9️⃣ Copy vs View — ⚠️ Important!

This is one of the most common sources of bugs in Pandas!

- **View**: A reference to the same data. Changing the view **changes the original**.
- **Copy**: An independent copy. Changing the copy does **NOT** affect the original.

> 🛑 Slicing (`head()`, `[]`) can return a view or a copy depending on context. Always use `.copy()` when you want independence.


````python
original = pd.Series([10, 20, 30, 40, 50])

# View (no .copy()) — modifying new_s WILL affect original
new_s = original.head(3)
new_s[0] = 999
print("original after modifying view:")
print(original)
````

````python
original = pd.Series([10, 20, 30, 40, 50])

# Copy — modifying new_s will NOT affect original
new_s = original.head(3).copy()
new_s[0] = 999
print("original after modifying copy:")
print(original)
````

## 1️⃣1️⃣ Broadcasting & Arithmetic Operators

Operations on a Series are applied **element-wise** — this is called **broadcasting**.
When operating between two Series, Pandas aligns them by **index labels**.


````python
s = pd.Series([10, 20, 30, 40])

print("s + 100:")
print(s + 100)         # adds 100 to every element

print("
s * 2:")
print(s * 2)           # doubles every element

print("
s ** 2:")
print(s ** 2)          # squares every element
````

````python
# Series + Series — aligned by index
a = pd.Series([1, 2, 3], index=['x', 'y', 'z'])
b = pd.Series([10, 20, 30], index=['x', 'y', 'z'])
print(a + b)
````

````python
# Mismatched index → NaN for unmatched labels
a = pd.Series([1, 2, 3], index=['x', 'y', 'z'])
b = pd.Series([10, 20, 30], index=['x', 'y', 'w'])  # 'z' vs 'w'
print(a + b)   # 'z' and 'w' become NaN
````

````python
# Relational operators — return boolean Series
s = pd.Series([10, 55, 30, 80, 45])
print(s >= 50)
````

## 1️⃣2️⃣ Boolean Indexing

Boolean indexing lets you **filter** a Series based on a condition. It's one of the most powerful and frequently used patterns in Pandas.

Pattern: `series[condition]`


````python
scores = pd.Series([45, 82, 33, 91, 60, 78, 55, 20])

# Elements > 60
print("scores > 60:")
print(scores[scores > 60])
````

````python
# How many elements satisfy condition
print("count > 60:", scores[scores > 60].size)
print("count > 60:", (scores > 60).sum())    # faster way!
````

````python
# Multiple conditions — use & (AND) and | (OR), wrap each in ()
print("between 50 and 80:")
print(scores[(scores >= 50) & (scores <= 80)])
````

````python
# NOT condition — use ~
print("NOT > 60 (i.e., <= 60):")
print(scores[~(scores > 60)])
````

## 1️⃣4️⃣ Important Series Methods

### `astype()` — Change Data Type
Convert the dtype of a Series to save memory or for compatibility.


````python
import sys
s = pd.Series([1, 2, 3, 4, 5])

print("Default dtype:", s.dtype)
print("Memory (int64):", sys.getsizeof(s), "bytes")

s_small = s.astype('int8')
print("
Converted dtype:", s_small.dtype)
print("Memory (int8):", sys.getsizeof(s_small), "bytes")
````

### `between()` — Range Filter
A cleaner way to filter values within a range (inclusive on both ends).


````python
scores = pd.Series([10, 45, 78, 52, 33, 90, 61])

# Find scores between 50 and 80
scores[scores.between(50, 80)]
````

### `clip()` — Cap Values
Replaces values **below** a lower bound or **above** an upper bound with those bounds.
Useful for removing outliers without dropping data.


````python
s = pd.Series([5, 150, 80, 300, 200, 50, 10])
print("Original:")
print(s)

print("
Clipped to [50, 200]:")
print(s.clip(50, 200))   # anything < 50 becomes 50; anything > 200 becomes 200
````

### `drop_duplicates()` & `duplicated()` — Handle Duplicates


````python
temp = pd.Series([1, 1, 2, 2, 3, 3, 4, 4])

print("drop_duplicates (keep first):")
print(temp.drop_duplicates())

print("
drop_duplicates (keep last):")
print(temp.drop_duplicates(keep='last'))

print("
duplication count:", temp.duplicated().sum())
print("duplicated() mask:")
print(temp.duplicated())
````

### `isnull()` / `notnull()` / `dropna()` / `fillna()` — Handling Missing Values

Missing values appear as `NaN` (Not a Number). This is one of the most critical skills in data cleaning.


````python
s = pd.Series([1, 2, np.nan, 4, np.nan, 6])

print("isnull:")
print(s.isnull())

print("
notnull:")
print(s.notnull())

print("
Missing value count:", s.isnull().sum())
print("Non-missing count:  ", s.count())
````

````python
# dropna — remove all NaN entries
print("After dropna:")
print(s.dropna())
````

````python
# fillna — replace NaN with a value or strategy
print("Fill with 0:")
print(s.fillna(0))

print("
Fill with mean:")
print(s.fillna(s.mean()))

print("
Forward fill (propagate last valid value):")
print(s.fillna(method='ffill'))

print("
Backward fill:")
print(s.fillna(method='bfill'))
````

### `isin()` — Check Membership
Filter elements that appear in a given list. Cleaner than chaining multiple `==` conditions.


````python
scores = pd.Series([0, 25, 49, 50, 75, 99, 100])

# Long way
print(scores[(scores == 49) | (scores == 99)])

# Clean way with isin()
print(scores[scores.isin([49, 99])])
````

### `apply()` — Apply a Function Element-wise
Apply any function (built-in, lambda, or custom) to each element of the Series.


````python
s = pd.Series([1, 2, 3, 4, 5])

# Using lambda
print(s.apply(lambda x: x ** 2))
````

````python
# Grade categorization
scores = pd.Series([45, 80, 55, 92, 33, 67])

def grade(x):
    if x >= 90: return 'A'
    elif x >= 75: return 'B'
    elif x >= 60: return 'C'
    elif x >= 45: return 'D'
    else: return 'F'

scores.apply(grade)
````

````python
# Apply on string series
names = pd.Series(['alice', 'bob', 'charlie'])
names.apply(lambda x: x.capitalize())
````

### `copy()` — Safe Copy
Always use `.copy()` when you want a truly independent copy of a Series.


````python
original = pd.Series([10, 20, 30])

safe_copy = original.copy()
safe_copy[0] = 999

print("original:", original.values)      # unchanged
print("safe_copy:", safe_copy.values)    # modified
````

## 1️⃣5️⃣ Additional Commonly Used Methods (Not Covered Above)

### `unique()` & `nunique()` — Unique Values


````python
s = pd.Series(['a', 'b', 'a', 'c', 'b', 'a'])

print("unique values:", s.unique())      # array of unique values
print("number of unique:", s.nunique())  # count of unique values
````

### `rename()` — Rename Index Labels


````python
s = pd.Series([67, 57, 89], index=['maths', 'english', 'science'])
s_renamed = s.rename({'maths': 'Math', 'english': 'Eng', 'science': 'Sci'})
print(s_renamed)
````

### `map()` — Map Values Through a Dictionary or Function
Similar to `apply()` but designed for **substituting** values using a mapping.


````python
grades = pd.Series(['A', 'B', 'C', 'A', 'D'])
grade_points = {'A': 4, 'B': 3, 'C': 2, 'D': 1}

grades.map(grade_points)
````

### `str` Accessor — String Operations
If your Series contains strings, you can use `.str.<method>()` for vectorized string operations.


````python
names = pd.Series(['alice smith', 'BOB JONES', '  Charlie  ', 'dave'])

print(names.str.upper())           # UPPER CASE
print(names.str.lower())           # lower case
print(names.str.title())           # Title Case
print(names.str.strip())           # Remove leading/trailing whitespace
print(names.str.contains('ali', case=False))  # Boolean mask
print(names.str.len())             # Length of each string
print(names.str.replace('dave', 'David'))     # Replace
print(names.str.split())           # Split into list
````

### `dt` Accessor — Datetime Operations
If your Series contains datetime values, use `.dt.<attribute>()` to extract parts.


````python
dates = pd.Series(pd.date_range('2024-01-01', periods=5, freq='M'))

print(dates.dt.year)
print(dates.dt.month)
print(dates.dt.day)
print(dates.dt.day_name())
print(dates.dt.is_month_end)
````

### `where()` and `mask()` — Conditional Replace
- `where(cond)` — keep values where condition is True, replace others with NaN (or fill value)
- `mask(cond)` — opposite of `where`


````python
s = pd.Series([10, 25, 5, 40, 15])

# Keep values > 15; replace others with NaN
print("where (keep >15):")
print(s.where(s > 15))

# Replace values > 15 with -1
print("
where with fill:")
print(s.where(s > 15, other=-1))

# mask — replace values WHERE condition IS TRUE
print("
mask (replace >15 with 0):")
print(s.mask(s > 15, other=0))
````

### `idxmin()` / `idxmax()` — Index of Min/Max Value
Returns the **index label** of the minimum or maximum value.


````python
s = pd.Series([30, 10, 50, 20, 40], index=['a','b','c','d','e'])

print("Index of min value:", s.idxmin())   # 'b'
print("Index of max value:", s.idxmax())   # 'c'
````

### `nlargest()` / `nsmallest()` — Top N Values
Much faster than sorting the entire Series when you only need the top/bottom N.


````python
s = pd.Series([30, 10, 50, 20, 40, 45, 5])

print("Top 3 largest:")
print(s.nlargest(3))

print("
Top 3 smallest:")
print(s.nsmallest(3))
````

### `rank()` — Rank Values
Assigns a rank to each element. Useful in competitive analysis.


````python
scores = pd.Series([70, 50, 90, 80, 60], index=['A','B','C','D','E'])

# Default: average rank for ties, ascending order
print(scores.rank())
print()
# Descending rank (rank 1 = highest)
print(scores.rank(ascending=False))
````

### `pct_change()` — Percentage Change
Computes the percentage change between consecutive elements. Very common in financial/time-series analysis.


````python
prices = pd.Series([100, 110, 105, 120, 115])
prices.pct_change() * 100   # multiply by 100 for % values
````

### `corr()` — Correlation Between Two Series
Computes the Pearson correlation coefficient (-1 to 1).


````python
a = pd.Series([1, 2, 3, 4, 5])
b = pd.Series([2, 4, 5, 4, 5])

print("Correlation:", a.corr(b))
````

### `reset_index()` — Reset the Index
Converts the index back to a default integer index (and turns the old index into a column if `drop=False`).


````python
s = pd.Series([10, 20, 30], index=['a','b','c'])

print("reset_index (drop=True):")
print(s.reset_index(drop=True))   # discard old index

print("
reset_index (drop=False):")
print(s.reset_index())            # old index becomes a column -> returns DataFrame
````
