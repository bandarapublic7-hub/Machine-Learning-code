## 📌 What is a DataFrame?

A **Pandas DataFrame** is a **2-dimensional labeled data structure** — essentially a table with:
- **Rows** (observations, records)
- **Columns** (variables, features)
- **Row index** (labels for rows)
- **Column names** (labels for columns)

Think of it as:
- A spreadsheet (Excel, Google Sheets)
- A SQL table
- A dictionary of Series objects (each column is a Series)

```
        Name    Age    Salary
    0   Alice   28    50000
    1   Bob     34    60000
    2   Carol   29    55000
```

> 💡 DataFrames are **the most important data structure in Pandas** — 90% of real-world work involves them!

## 2️⃣ Creating DataFrames

### 📂 From a List of Lists

When you pass a 2D list (list of lists), Pandas creates a DataFrame with **auto-generated column names** (0, 1, 2, ...).


````python
student_data = [
    [100, 80, 10],
    [90, 70, 7],
    [120, 100, 14],
    [80, 50, 2]
]

df = pd.DataFrame(student_data, columns=['IQ', 'Marks', 'Package'])
df
````

### 📂 From a Dictionary

When you create a DataFrame from a **dict**, the **keys become column names** and **values become columns** (as lists or Series).

## 5️⃣ DataFrame — Mathematical Methods

DataFrames have powerful aggregation methods. **The `axis` parameter is crucial!**

- **`axis=0`** (default) — compute **down columns** (returns Series with one value per column)
- **`axis=1`** — compute **across rows** (returns Series with one value per row)

| Method | Description |
|--------|-------------|
| `.sum()` | Sum (axis parameter matters!) |
| `.mean()` | Average |
| `.median()` | Median |
| `.std()` | Standard deviation |
| `.var()` | Variance |
| `.min()`, `.max()` | Minimum, Maximum |
| `.count()` | Non-null count (useful with missing data) |
| `.cumsum()` | Cumulative sum |
| `.cumprod()` | Cumulative product |
| `.abs()` | Absolute value |
| `.round(n)` | Round to n decimals |
| `.rank()` | Rank values |


````python
df = pd.DataFrame({
    'Math': [85, 92, 78, 95, 88],
    'Science': [90, 88, 92, 89, 91],
    'English': [78, 95, 88, 92, 85]
}, index=['Student1', 'Student2', 'Student3', 'Student4', 'Student5'])

print("Sum (axis=0, sum each COLUMN):")
print(df.sum(axis=0))

print("
Sum (axis=1, sum each ROW = total score):")
print(df.sum(axis=1))
````

````python
print("Mean score per subject (axis=0):")
print(df.mean(axis=0))

print("
Mean score per student (axis=1):")
print(df.mean(axis=1))
````

````python
print("Variance:")
print(df.var())

print("
Standard Deviation:")
print(df.std())
````

````python
print("Min and Max:")
print("Min:", df.min().to_dict())
print("Max:", df.max().to_dict())
````

````python
# cumsum — running total by column
print("Cumulative sum:")
print(df.cumsum())
````

## 7️⃣ Selecting Rows

Use **`.iloc[]`** for **integer position-based** indexing (like Python lists).
Use **`.loc[]`** for **label-based** indexing (by row index).

> ✅ **Best practice:** Always prefer `.iloc` and `.loc` over `.[]` for clarity!

| Syntax | Meaning |
|--------|---------|
| `df.iloc[0]` | First row (by position) |
| `df.iloc[[0,2,4]]` | Rows at positions 0, 2, 4 |
| `df.iloc[0:3]` | Rows 0-2 (0:3 excludes 3) |
| `df.loc['name']` | Row labeled 'name' |
| `df.loc[['n1', 'n2']]` | Rows labeled 'n1', 'n2' |
| `df.loc['a':'c']` | Rows 'a' to 'c' (inclusive!) |


````python
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Carol', 'David', 'Eve'],
    'Age': [25, 30, 28, 35, 22],
    'Salary': [50000, 60000, 55000, 70000, 48000]
})

# Single row by position — returns SERIES
print("iloc[0] (first row):")
print(df.iloc[0])
print()

# Multiple rows by position
print("iloc[0:3] (rows 0,1,2):")
print(df.iloc[0:3])
print()

# Fancy indexing
print("iloc[[1, 3]] (rows 1 and 3):")
print(df.iloc[[1, 3]])
````

````python
# With custom index
df_indexed = df.set_index('Name')
print("DataFrame with Name as index:")
print(df_indexed)
print()

# loc — label-based
print("loc['Alice']:")
print(df_indexed.loc['Alice'])
print()

print("loc[['Alice', 'David']]:")
print(df_indexed.loc[['Alice', 'David']])
````

````python
# loc slicing is INCLUSIVE on both ends (different from iloc!)
print("loc['Alice':'Carol'] (inclusive):")
print(df_indexed.loc['Alice':'Carol'])
````

## 8️⃣ Selecting Rows AND Columns

Use **`.iloc[rows, cols]`** or **`.loc[rows, cols]`** to select both dimensions at once.

Syntax: `df.iloc[row_indexer, col_indexer]`


````python
df = pd.DataFrame({
    'A': [1, 2, 3, 4, 5],
    'B': [10, 20, 30, 40, 50],
    'C': [100, 200, 300, 400, 500],
    'D': [1000, 2000, 3000, 4000, 5000]
})

# First 3 rows, first 2 columns
print("iloc[0:3, 0:2] (rows 0-2, cols 0-1):")
print(df.iloc[0:3, 0:2])
print()

# All rows, columns A and C
print("df.loc[:, ['A', 'C']]:")
print(df.loc[:, ['A', 'C']])
print()

# Rows 1 and 3, columns B and D
print("iloc[[1,3], [1,3]]:")
print(df.iloc[[1, 3], [1, 3]])
````

## 9️⃣ Boolean Indexing (Filtering)

Filter rows based on **conditions**. This is one of the **most important patterns** in Pandas!

Pattern: `df[condition]` or `df.loc[condition]`

> ⚠️ Use `&` (AND), `|` (OR), `~` (NOT). Do NOT use `and`, `or`, `not` keywords!


````python
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Carol', 'David', 'Eve'],
    'Age': [25, 30, 28, 35, 22],
    'Salary': [50000, 60000, 55000, 70000, 48000]
})

# Single condition
print("Age > 28:")
print(df[df['Age'] > 28])
print()

# Multiple conditions (AND)
print("Age > 28 AND Salary > 55000:")
print(df[(df['Age'] > 28) & (df['Salary'] > 55000)])
````

````python
# OR condition
print("Age < 25 OR Salary > 65000:")
print(df[(df['Age'] < 25) | (df['Salary'] > 65000)])
print()

# NOT condition
print("NOT (Age > 30):")
print(df[~(df['Age'] > 30)])
````

````python
# String filtering
print("Name contains 'a' (case-insensitive):")
print(df[df['Name'].str.lower().str.contains('a')])
print()

# isin — check membership
print("Name in ['Alice', 'David']:")
print(df[df['Name'].isin(['Alice', 'David'])])
````

````python
# Get count without viewing all rows
print("How many earn > 55000?")
print((df['Salary'] > 55000).sum())
````

## 1️⃣4️⃣ Data Cleaning

### Missing Values

This is one of the **most critical skills** in data science!


````python
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', None, 'David', 'Eve'],
    'Age': [25, None, 30, 35, 22],
    'Salary': [50000, 60000, np.nan, 70000, 48000]
})

print("Original DataFrame:")
print(df)
print()

print("Missing value count:")
print(df.isnull().sum())
````

````python
print("\nRows with at least one missing value:")
print(df[df.isnull().any(axis=1)])
````

````python
# Drop rows with any missing value
print("\nDrop rows with ANY null:")
print(df.dropna())
print()

# Drop rows where ALL values are missing
print("Drop rows where ALL are null (rare):")
print(df.dropna(how='all'))
````

````python
# Fill missing values
print("\nFill with 0:")
print(df.fillna(0))
print()

print("Fill with forward fill (propagate last valid value):")
print(df.fillna(method='ffill'))
````

````python
# Fill with mean/median/mode
df_copy = df.copy()
df_copy['Salary'] = df_copy['Salary'].fillna(df_copy['Salary'].mean())
print("\nFill Salary with mean:")
print(df_copy)
````

## 1️⃣5️⃣ Data Type Conversion

### astype()
Convert column data types. Critical for memory efficiency and compatibility.


````python
df = pd.DataFrame({
    'ID': [1, 2, 3, 4, 5],
    'Name': ['A', 'B', 'C', 'D', 'E'],
    'Score': [85.5, 92.0, 78.5, 95.0, 88.5]
})

print("Original dtypes:")
print(df.dtypes)
print()

# Convert to int32 (save memory)
df['ID'] = df['ID'].astype('int32')

# Convert to string
df['Name'] = df['Name'].astype('string')

# Convert to category (for repetitive values)
df['Grade'] = pd.cut(df['Score'], bins=[0, 70, 85, 100], labels=['C', 'B', 'A']).astype('category')

print("\nNew dtypes:")
print(df.dtypes)
print()

print("DataFrame:")
print(df)
````

````python
# Check memory usage
print("Memory usage (original):", pd.DataFrame({'ID': [1,2,3,4,5]})['ID'].astype('int64').memory_usage(deep=True), "bytes")
print("Memory usage (int32):", pd.DataFrame({'ID': [1,2,3,4,5]})['ID'].astype('int32').memory_usage(deep=True), "bytes")
````

## 1️⃣6️⃣ GroupBy — One of the Most Powerful Operations

`groupby()` lets you **split data into groups**, apply **aggregation functions**, and **combine results**.

This is essential for **data summarization and analysis**.

Pattern: `df.groupby('column').aggregation_function()`


````python
df = pd.DataFrame({
    'Department': ['Sales', 'Sales', 'IT', 'IT', 'HR', 'HR'],
    'Name': ['Alice', 'Bob', 'Carol', 'David', 'Eve', 'Frank'],
    'Salary': [50000, 55000, 70000, 75000, 45000, 48000]
})

print("Original:")
print(df)
print()

# Group by Department and sum salaries
print("Total salary by Department:")
print(df.groupby('Department')['Salary'].sum())
````

````python
# Multiple aggregations
print("\nDepartment statistics:")
print(df.groupby('Department')['Salary'].agg(['count', 'mean', 'min', 'max']))
````

````python
# Group by multiple columns
df2 = pd.DataFrame({
    'City': ['NYC', 'NYC', 'LA', 'LA', 'NYC', 'LA'],
    'Department': ['Sales', 'IT', 'Sales', 'IT', 'Sales', 'IT'],
    'Salary': [50000, 70000, 55000, 75000, 52000, 72000]
})

print("\nGrouped by City and Department:")
print(df2.groupby(['City', 'Department'])['Salary'].mean())
````

## 1️⃣7️⃣ Merging DataFrames

Combine DataFrames based on a **common key**, similar to SQL JOIN.

| Type | Syntax | Description |
|------|--------|-------------|
| Inner Join | `merge(..., how='inner')` | Only matching keys |
| Left Join | `merge(..., how='left')` | All from left, matching from right |
| Right Join | `merge(..., how='right')` | All from right, matching from left |
| Outer Join | `merge(..., how='outer')` | All keys from both |
| Concatenate | `pd.concat([df1, df2])` | Stack vertically or horizontally |


````python
# Sample data
students = pd.DataFrame({
    'StudentID': [1, 2, 3, 4],
    'Name': ['Alice', 'Bob', 'Carol', 'David']
})

scores = pd.DataFrame({
    'StudentID': [1, 2, 3, 4],
    'Math': [85, 92, 78, 95],
    'Science': [90, 88, 92, 89]
})

print("Students:")
print(students)
print()
print("Scores:")
print(scores)
print()

# Inner merge
merged = pd.merge(students, scores, on='StudentID', how='inner')
print("Merged (inner join):")
print(merged)
````

````python
# Concatenate vertically (stack rows)
df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})

print("\nConcatenate rows:")
print(pd.concat([df1, df2], ignore_index=True))
````

````python
# Concatenate horizontally (stack columns)
print("\nConcatenate columns:")
print(pd.concat([df1, df2], axis=1))
````

## 1️⃣8️⃣ Pivot Tables

Reshape data for **cross-tabulation and summarization** — very useful for data analysis!


````python
df = pd.DataFrame({
    'City': ['NYC', 'NYC', 'LA', 'LA', 'NYC', 'LA'],
    'Product': ['A', 'B', 'A', 'B', 'A', 'B'],
    'Sales': [100, 150, 120, 200, 110, 180]
})

print("Original:")
print(df)
print()

# Pivot table
pivot = df.pivot_table(values='Sales', index='City', columns='Product', aggfunc='sum')
print("Pivot Table:")
print(pivot)
````

## 1️⃣9️⃣ String Operations using `.str` Accessor

DataFrames with string columns can use the `.str` accessor for **vectorized string operations**.


````python
df = pd.DataFrame({
    'Name': ['alice johnson', 'BOB SMITH', '  CAROL WHITE  '],
    'Email': ['alice@gmail.com', 'bob@yahoo.com', 'carol@gmail.com']
})

print("Original:")
print(df)
print()

# String methods
print("Uppercase names:")
print(df['Name'].str.upper())
print()

print("Stripped and title case:")
print(df['Name'].str.strip().str.title())
print()

print("Extract domain from email:")
print(df['Email'].str.extract(r'@(.+)')[0])
````

## 2️⃣0️⃣ Datetime Operations using `.dt` Accessor

Work with dates and times efficiently.


````python
df = pd.DataFrame({
    'Date': pd.date_range('2024-01-01', periods=5),
    'Sales': [100, 150, 120, 200, 110]
})

print("Original:")
print(df)
print()

# Extract components
print("Year:", df['Date'].dt.year.tolist())
print("Month:", df['Date'].dt.month.tolist())
print("Day:", df['Date'].dt.day.tolist())
print("Day Name:", df['Date'].dt.day_name().tolist())
````

## 2️⃣1️⃣ Apply — Row-wise or Column-wise Functions

Apply a function to each row or column.


````python
df = pd.DataFrame({
    'Math': [85, 92, 78, 95],
    'Science': [90, 88, 92, 89],
    'English': [78, 95, 88, 92]
})

# Apply to each row (axis=1) — calculate total score
df['Total'] = df.apply(lambda row: row.sum(), axis=1)
print("After adding Total:")
print(df)
print()

# Apply to each column (axis=0) — find mean
print("Mean of each subject:")
print(df.apply(lambda col: col.mean(), axis=0))
````

## 2️⃣2️⃣ More Important Methods

### value_counts()
Count unique values in a Series or DataFrame column.


````python
df = pd.DataFrame({
    'Fruit': ['Apple', 'Banana', 'Apple', 'Orange', 'Banana', 'Banana'],
    'Color': ['Red', 'Yellow', 'Green', 'Orange', 'Yellow', 'Yellow']
})

print("Fruit counts:")
print(df['Fruit'].value_counts())
print()

print("As percentage:")
print(df['Fruit'].value_counts(normalize=True) * 100)
````

### unique() and nunique()


````python
print("Unique fruits:", df['Fruit'].unique())
print("Number of unique fruits:", df['Fruit'].nunique())
````

### nlargest() and nsmallest()


````python
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Carol', 'David', 'Eve'],
    'Salary': [50000, 75000, 60000, 80000, 55000]
})

print("Top 3 highest salaries:")
print(df.nlargest(3, 'Salary'))
print()

print("Top 2 lowest salaries:")
print(df.nsmallest(2, 'Salary'))
````

### reset_index()
Convert the index back to a default integer index.


````python
df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6]
}, index=['x', 'y', 'z'])

print("Original with custom index:")
print(df)
print()

print("After reset_index:")
print(df.reset_index())
````

### set_index()
Make a column the new index.


````python
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Carol'],
    'Age': [25, 30, 28],
    'Salary': [50000, 60000, 55000]
})

print("Set Name as index:")
print(df.set_index('Name'))
````
